```python
import seaborn as sns
sns.set(style='white', color_codes=True)
import warnings # scipy/stats/stats.py error with seaborn
warnings.filterwarnings('ignore')
```

```python
sns.pairplot(iris.drop(iris.columns[5:iris.shape[1]], axis=1), markers=["o","D","s"], hue='Species', size=3, diag_kind='hist')
```

**Setosa is being the intruder**

```python
sns.violinplot(x='Species',y='Petal.Length', data=iris, size=6)
```

**It's pretty clear that Setosa is different from virginica and versicolor**

```python
sns.pairplot(iris.drop(iris.columns[5:iris.shape[1]], axis=1).drop(iris.index[:50]), hue='Species', size=3, diag_kind='hist', kind='reg')
```

**We could separate virginica and versicolor by *Petal.Length/Sepal.Width* and *Petal.Width/Sepal.Width***
